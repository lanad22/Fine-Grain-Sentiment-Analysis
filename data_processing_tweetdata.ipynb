```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data-tweets/twitter-2016train-CE.tsv', sep='\t', header=None, names=['user', 'at', 'label', 'text'])
# Convert the label column to a categorical type
df['label'] = df['label'].astype(int).astype('category')

# Set display option for pandas
pd.set_option('display.max_colwidth', None)

# Get 5 examples of each label
examples_per_label = df.groupby('label').apply(lambda x: x.sample(5, random_state=1)).reset_index(drop=True)

# Display the result
examples_per_label
```

```python
label_counts = df['label'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
label_counts.plot(kind='barh')
plt.title('Distribution of Labels Train Set')
plt.xlabel('Label')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
df_test = pd.read_csv('data-tweets/twitter-2016test-CE.tsv', sep='\t', header=None, names=['user', 'at', 'label', 'text'])
# Convert the label column to a categorical type
df_test['label'] = df_test['label'].astype(int).astype('category')

label_counts_test = df_test['label'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
label_counts_test.plot(kind='barh',color="#8B0000")
plt.title('Distribution of Labels Test Set')
plt.xlabel('Label')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```
